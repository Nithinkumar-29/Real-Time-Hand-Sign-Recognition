# Creating Dataset

### 1.Gathering Images

```python
import cv2 as cv
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
# mp_drawing,mp_hands
```

```python
c = cv.VideoCapture(0)
i1 = 0;i2=0;i3=0;i4=0;i5=0
while True:
    r,f = c.read()
    img = cv.flip(f,2)
    cv.imshow("x",img)
    if(cv.waitKey(10)==ord('a')):
        break
    if(cv.waitKey(10)==ord('1')):
        i1 = i1+1
        cv.imwrite("1/"+str(i1)+".jpg",img)
    if(cv.waitKey(10)==ord('2')):
        i2 = i2+1
        cv.imwrite("2/"+str(i2)+".jpg",img)
    if(cv.waitKey(10)==ord('3')):
        i3 = i3+1
        cv.imwrite("3/"+str(i3)+".jpg",img)
    if(cv.waitKey(10)==ord('4')):
        i4 = i4+1
        cv.imwrite("4/"+str(i4)+".jpg",img)
    if(cv.waitKey(10)==ord('5')):
        i5 = i5+1
        cv.imwrite("5/"+str(i5)+".jpg",img)
c.release()
cv.destroyAllWindows()
```

### 2.Creating csv file

```python
import pandas as pd
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
# mp_drawing,mp_hands
```

```python
x = enumerate(mp_hands.HandLandmark)
col_names = ['label']
for i,j in x:
    col_names.append(str(j).removeprefix('HandLandmark.')+"_x")
    col_names.append(str(j).removeprefix('HandLandmark.')+"_y")
    col_names.append(str(j).removeprefix('HandLandmark.')+"_z")
```

```python
col_names
```

```python
a = dict()
for i in col_names:
    a.update({i:'x'})
```

```python
a
```

```python
df = pd.DataFrame(a,index=[0])
df
```

```python
df = df.drop(index=0,axis=0)
df
```

```python
hands = mp_hands.Hands(max_num_hands=2,min_detection_confidence=0.7,min_tracking_confidence=0.5)
for i in range(1,6):
    path = 'data/'+str(i)
#     lim = 0
    for j in os.listdir(path):
        if(j.endswith('.jpg')):
            index = 0
            img_path = path+'/'+j
            img = cv.imread(img_path)
            img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
            img.flags.writeable = False
            result = hands.process(img)
            img.flags.writeable = True
            img = cv.cvtColor(img,cv.COLOR_RGB2BGR)
            if result.multi_hand_landmarks:
                for hand in result.multi_hand_landmarks:
                    l = [i]
#                     print(hand.landmark)
                    l1 = list(hand.landmark)
                    l2 = []
                    for val in l1:
                        l2.append(val.x)
                        l2.append(val.y)
                        l2.append(val.z)
                    l+=l2
#                     print(l)
                    row = dict()
                    for q in range(0,len(col_names)):
                        row[col_names[q]] = l[q]
#                     print(row)
                    df = df.append(row,ignore_index=True)
#         lim = 1
```
