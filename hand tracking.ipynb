### Building the model

```python
import cv2 as cv
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pyautogui
import time
import pandas as pd
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
# mp_drawing,mp_hands
```

```python
import pandas as pd
import os
```

```python
df = pd.read_csv("dataset-1.csv")
df
```

```python
ex_df = df
ex_df['label'] = df['label'].replace(5,0)
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pyautogui
import time
```

```python
X = ex_df.drop('label',axis=1)
y = ex_df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape,y_train.shape
```

```python
X_train = np.asarray(X_train).astype('float32')
y_train = np.asarray(y_train).astype('float32')
X_test = np.asarray(X_test).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
# from sklearn.preprocessing import S
# scaler = StandardScaler()
# X_train = scaler.fit_transform(np.array(X_train).reshape(-1,1))
# X_test = scaler.transform(np.array(X_test).reshape(-1,1))
# y_train = scaler.fit_transform(np.array(y_train).reshape(-1,1))
# y_test = scaler.transform(np.array(y_test).reshape(-1,1))
```

```python
X_train.shape,y_train.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(63,activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(5,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=1000)
```

```python
y_pred = model.predict(X_test)
model.evaluate(X_test,y_test)
y_pred
```

```python
model.save('handgesture.hdf5')
```

### Testing with live cam

```python
hands = mp_hands.Hands(max_num_hands=2,min_detection_confidence=0.7,min_tracking_confidence=0.5)
c = cv.VideoCapture(0)
while True:
    b,f = c.read()
    #BGR 2 RGB
    img = cv.cvtColor(f,cv.COLOR_BGR2RGB)
    
    #flip horizontally
    img = cv.flip(img,1)
    
    #making image not writable
    img.flags.writeable=False
    
    #detect
    result = hands.process(img)
    
    #make image writable
    img.flags.writeable=True
    
    #RBG 2 BGR
    img = cv.cvtColor(img,cv.COLOR_RGB2BGR)
    
    #draw results
    if result.multi_hand_landmarks:
        for hand in result.multi_hand_landmarks:
            l1 = list(hand.landmark)
            l2 = []
            for val in l1:
                l2.append(val.x)
                l2.append(val.y)
                l2.append(val.z)
            l2 = np.asarray(l2).astype('float32')
            l2 = l2.reshape((1,1,63))
            label = m.predict(l2)
            label = np.argmax(np.squeeze(label))
            if(label==0):
                label = 5
#                 pyautogui.press('c')
#             if(label==1):
#                 pyautogui.press('a',presses=10)
#             if(label==2):
#                 pyautogui.press('d',presses=10)
#             if(label==3):
#                 pyautogui.press('x')
#             if(label==4):
#                 pyautogui.press('z')
            text = "Label: "+str(label)
            center_x = int(hand.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x*img.shape[1])
            center_y = int(hand.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y*img.shape[0])
            rect_x = int(hand.landmark[mp_hands.HandLandmark.THUMB_TIP].x*img.shape[1])
            rect_y = int(hand.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y*img.shape[0])
            wrist_x = int(hand.landmark[mp_hands.HandLandmark.PINKY_TIP].x*img.shape[1])
            wrist_y = int(hand.landmark[mp_hands.HandLandmark.WRIST].y*img.shape[0])
            center = (center_x,center_y)
            rect = (rect_x,rect_y)
            wrist = (wrist_x,wrist_y)
            mp_drawing.draw_landmarks(img,hand,mp_hands.HAND_CONNECTIONS,
            mp_drawing.DrawingSpec((0,255,255)),
            mp_drawing.DrawingSpec((160,30,240))
            )
            img = cv.circle(img,center,15,(255,0,0))
            img = cv.rectangle(img,rect,wrist,(0,255,0))
            img = cv.putText(img,text,(rect_x-50,rect_y-50),cv.FONT_HERSHEY_PLAIN,1,(0,0,255),2)
    #displaying the results
    img = cv.resize(img,(0,0),fx=2,fy=2)
    cv.imshow("tracking yo hands",img)
    if(cv.waitKey(5)==ord('e')):
        break
c.release()
cv.destroyAllWindows()
```
